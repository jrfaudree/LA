# Vectors

## Vectors of all zeros or all ones

The functions `zeros` and `ones` give you vectors of all zeros or all ones.

````julia
zeros(4)
````

````julia
ones(7)
````

## Vectors of evenly spaced numbers

It's common to make a vector where the first entry is some $a_0$, the last entry is $a_n$, and all the numbers in between are evenly spaced.  For example, you might want times starting at 0 seconds, ending at 1.5 seconds, with each intermediate entry separated by 0.1 seconds.  You do this as follows:

```
collect( 0:0.1:1)
```

The part `0:0.1:1.5` creates an object that records the beginning, ending, and step size.

````julia
0:0.1:1.5
````

The `collect` part turns that information into a vector.

````julia
collect(0:0.1:1.5)
````

A related task is to make a vector where you specify the first and last element, but you specify the length of the vector rather than the step size; you let Julia figure out the step size.

For example, to make a list of 10 numbers starting at 0 and ending at $2\pi$ you enter:


````julia
collect(range(0, 2*pi, 10))
````

## Entering your own vectors

Use square brackets and list the numbers making up the vector, separated by commas.

````julia
x = [ 4.3, 9.7, pi]
````

Although we won't talk about matrices until Chapter 6, there are some things you need to know about matrices because they affect you input vectors in Julia.

If you forget the commas, this is what you get instead:

````julia
[4.3 9.7 pi]
````

That's not a vector, per-se.  It's a $1\times 3$ matrix, a matrix with one row and three columns. 

Rows in a matrix are separated by semicolons.

````julia
[ 4.3 9.7 pi; 4 5 6]
````

Or you can simply enter each row on its own separate line.

````julia
[4.3 9.7 pi;
    4 5 6 ]
````

The key things to remember:
- Use commas to separate entries of a vector
- Use spaces to separate entries in a row of a matrix
- Use semicolons to separate rows in a matrix

Sometimes vectors are identified with matrices with one column.  So Julia will let you use semicolons to separate the entries of a vector as well.

````julia
x_matrix = [4.3; 9.7; pi]
````

Computers use two different kinds of representations of numbers. 

1. Floating point numbers which represent real numbers like 2/3 and $\pi$ and $\sqrt{2}$.
2. Integers, which only represent, well, integers, like -27 and 6 and 0.

The `Float64` you saw in the previous example indicates the vector is made up of floating point numbers.  Julia will make a vector of integers if it can instead, in which case you'll see `Int64` instead of `Float64`.  You can mostly ignore this.

````julia
x_int = [ -27, 6, 0 ]
````

But if you want, you can be explicit like this:

````julia
x_float = Float64[ -27, 6, 0]
````

Or just make sure at least one entry is not an integer:

````julia
x_float = [-27.0, 6, 0]
````

## Finding the length of a vector

Use the `length` function.  You might also see the `size` function, which works for matrices and higher-indexed arrays as well.

````julia
x = [ 1.3, 2.7, 9.6, -3.4]
````

````julia
length(x)
````

You might also see the `size` function, which is also used for matrices and other higher indexed objects.  It returns a list of numbers (a "tuple"), one for each array dimension (this will make more sense when we talk about matrices below.  For vectors, you'll see a tuple with one entry:


````julia
size(x)
````

You extract the first entry like this:

````julia
size(x)[1]
````

## Setting/retrieving entries of a vector

Use `x[k]` to reference the `k`th entry of a vector, which you can either retrieve or set as follows.

````julia
x = [4.3, 9.7, 18.2, -9.6]
````

````julia
x[1]
````

````julia
x[4]
````

## Obtaining a subvector

The expression `x[4:7]` refers to the a vector with entries `[x[4], x[5], x[6], x[7]]`, just like the notation $x_{4:7}$ from the text.

````julia
x = [1.3, -6.7, 4.3, -23.6, -4]
````

````julia
x[2:4]
````

## Stacking vectors to make a longer one

If `x` and `y` are vectors, then `[x; y]` is the vector obtained from tacking `y` onto the end of `x`.

````julia
x = [ 1, 2, 3, 4]
y = [7, 8, 9]

[x; y]
````

## Adding an element to the end of a vector

Use the `push!` command.  The `push` part comes from the idea of "pushing" another entry onto the end, and the exclamation mark is a Julia convention that one of the arguments (the vector in this case) is being modified.  Here it is in action.

````julia
x = [1.3, 3.9, 4.7]
````

````julia
push!(x, 29.5)
````

The vector `x` really did get modified:

````julia
x
````

# Vector Operations

## Vector addition and scalar multiplication

The fundamental vector operations are vector addition and scalar multiplication.  Use `+` and `*` for these.

````julia
x = [ 1.3, 2.9, 7.6]
y = [ 4.7, -1.9, -3.6];
````

````julia
x+y
````

````julia
3*x
````

````julia
0*x
````

## Elementwise multiplication

Multiplying two vectors is not a well defined operation. You'd think, maybe, this should be just multiplying each pair of entries (elementwise), but this turns out to be a less useful operation than the kinds of multiplicaiton we'll see in Chapter 6.  If you try to do it in Julia, you'll see an error message.

````julia
x = [ 1.3, 2.9, 7.6]
y = [ 4.7, -1.9, -3.6];
````

````julia
x*y
````

If you really want to do element-wise multiplication, you can use `.*` (like in MATLAB, if you are familiar with it).

````julia
x.*y
````

## Dot product

The book uses the notation $x^T y$ for what it calls the dot product.  In Julia, you write this as follows.`

````julia
x' * y
````

The prime on `x'` means the same thing as $T$.  It's a transpose operation, which will make more sense in Chapter 6 once we discuss matrices.

Order matters.  Look at this craziness:

````julia
y * x'
````

Again, we'll learn the big picture and the rules in Chapter 6.

Alternatively, to compute a dot product you can use the `dot` function, but only if the `LinearAlgebra` package has been loaded.

````julia
using LinearAlgebra
dot(x, y)
````

## Applying a function to every entry of a vector.

Sometimes you just want to apply a function, like $\sin$ to every entry of a vector.  Your first guess about how to do this might surprise you.

````julia
t = collect(range(0,2*pi, 10));
````

````julia
sin(t)
````

The problem is that `sin` eats numbers, not arrays of numbers.  Some languages, like MATLAB and Python extend `sin` so that it applys entrywise to vectors.  Julia uses a different approach. If you want to apply a function to every element of a vector, use a period (dot) before the parentheses, like this:

````julia
sin.(t)
````

## How to make surface and contour plots

A surface plot visualizes a function of two variables
by plotting $z=f(x,y)$.

A contour plot is an alternative way of visualizing a function of two variables and is analogous to
making a topographical map.  

We're going to work with the following function:
$$
f(x,y) = x^2 - y^2
$$

The graph of this function is known as a hyperboloid. 

### Surface plot

````julia
import Plots

function hyperboloid(x,y)
    return x^2 - y^2
end

# List of x coordinates
x = range(-2,2,100)

# List of y coordinates
y = range(-3,3,200)

# This makes a matrix of z coordinates.
#
# The entry Z[i,j] is f(x_j, y_i) where 
# x_j and y_i come from the vectors `x` and `y`
#
# The magic here comes from the @. 
#
# Don't worry too much about how this works, just
# use this as a recipe!
Z = @. hyperboloid(x', y)

Plots.surface(x, y, Z, xlabel="x", ylabel="y", zlabel="z", title="A hyperboloid")
````

### Contour plot

In a contour plot, each line is a line of constant elevation.  The function has the same value along each line.  For example, on the straight lines that intersect at the origin the function value is zero.

````julia
import Plots

function hyperboloid(x,y) 
    return x^2 - y^2
end

# Same technique as for the surface plot to make
# vectors for x and y coordinates and then a
# matrix Z that records the function values.
x = range(-2,2,length=100)
y = range(-2,2,length=100)

z = @. hyperboloid(x', y)

Plots.contour(x, y, z, color=:turbo,
    xlabel="x", ylabel="y", levels= 11, fill=true )
````
